```
import os
import zipfile
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))
zipf = zipfile.ZipFile('/content/YoloV3-EVA5_S13_executed.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('/content/YoloV3-EVA5_S13', zipf)
zipf.close()
```
